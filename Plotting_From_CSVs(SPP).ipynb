```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
stock_name= 'AAPL'
stock_location = '/content/drive/MyDrive/master_thesis/Test_Results/'+stock_name
result_detail_file = stock_location + '/result_detail_e1.csv'
result_overall_file = stock_location + '/result_overall.csv'
```

```python
def plot_from_csv(csv_file, csv_overall_file, title):
    year, month, day, episodes, step, earning_ratio, expected_assets, given_money, available_cash, stocks, n_holds, n_buys, n_sells, profits, commission_fee, sum, close, ep_reward, actions, eps, buy_signal, sell_signal, signal_average, loss = np.genfromtxt(csv_file, delimiter=",", unpack=True)
    year_o, month_o, day_o, episodes_o, step_o, earning_ratio_o, expected_assets_o, given_money_o, available_cash_o, stocks_o, n_holds_o, n_buys_o, n_sells_o, profits_o, commission_fee_o, sum_o, close_o, ep_reward_o, actions_o, eps_o, loss_o = np.genfromtxt(csv_overall_file, delimiter=",", unpack=True)
    num_epochs = 4
    
    markers_on_buy_hold = [buy_signal == 1]
    markers_on_buy = [actions == 1]
    markers_on_sell = [sell_signal == 1]
    
    markers_on_buy_signal = [signal_average == 1]
    markers_on_sell_signal = [signal_average == -1]

    plt.figure(1, figsize=(80,25))
    
    plt.plot(close, color='C0', lw=2.)
    plt.plot(close, '^', markersize=10, color='m', label='buy', markevery = markers_on_buy)
    plt.plot(close, 'v', markersize=10, color='k', label='sell', markevery = markers_on_sell)
    plt.title(title+'detail')
    #plt.plot(episodes, train_loss, label='Training Loss')
    #plt.plot(epoch, val_loss, label='Validation Loss')
    plt.legend()
    plt.xlabel('Episodes')
    #plt.xticks(np.arange(1, num_epochs + 1, step=1))
    plt.ylabel('Stock Price (Close)')
    plt.savefig(stock_location+'/Figures/'+title + 'detail.png')

    plt.figure(5, figsize=(15,5))
    plt.plot(stocks, color='b', lw=2.)
    plt.legend()
    plt.xlabel('Episodes')
    plt.ylabel('Stocks')
    plt.title(title + 'detail_Stocks')
    plt.savefig(stock_location+'/Figures/'+title + 'detail_stocks.png')


    plt.figure(2, figsize=(15,5))
    plt.plot(loss, color='b', lw=2.)
    #plt.plot(epoch, train_acc, label='Training Accuracy')
    #plt.plot(epoch, val_acc, label='Validation Accuracy')
    plt.legend()
    plt.xlabel('Episodes')
    #plt.xticks(np.arange(1, num_epochs + 1, step=1))
    plt.ylabel('Loss')
    plt.title(title + 'detail_Loss')
    plt.savefig(stock_location+'/Figures/'+title + 'detail_Loss.png')

    plt.figure(3, figsize=(15,5))
    plt.plot(earning_ratio_o, label='Earning Ratio')
    plt.legend()
    plt.xticks(np.arange(1, num_epochs+1, step=1))
    plt.xlabel('Epochs')
    plt.ylabel('Earning Ratio')
    plt.title(title + '_Overall_Earning_Ratio')
    plt.savefig(stock_location+'/Figures/'+title + '_Overall_Earning_Ratio.png')

    plt.figure(4, figsize=(15,5))
    plt.plot(loss_o, label='Loss')
    plt.legend()
    plt.xlabel('Epochs')
    plt.xticks(np.arange(1, num_epochs+1, step=1))
    plt.ylabel('Loss')
    plt.title(title + '_Overall_Loss')
    plt.savefig(stock_location+'/Figures/'+title + '_Overall_Loss.png')
```

```python
plot_from_csv(result_detail_file, result_overall_file, stock_name)
```
